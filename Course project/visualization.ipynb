```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Yulu.csv')
```

```python
# Преобраование datetime из строки в объект datetime
data['datetime'] = pd.to_datetime(data['datetime'])
```

```python
# Создадим новые столбцы для 'hour' и 'day of week'
data['hour'] = data['datetime'].dt.hour
data['day_of_week'] = data['datetime'].dt.dayofweek
```

```python
# Столбчатые диаграммы для категориальных признаков
cat_vars = ['season', 'holiday', 'workingday', 'weather']

for var in cat_vars:
    sns.countplot(x=var, data=data)
    plt.title(f'Распределение {var}')
    plt.show()
```

```python
# Тренд проката велосипедов по часам и дням недели
plt.figure(figsize=(12,6))
sns.pointplot(x='hour', y='count', hue='day_of_week', data=data)
plt.title('Тренд проката электросамокатов по часам и дням недели')
plt.show()
```

```python
# Heatmap для визуализации корреляции между признаками
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap корреляции')
plt.show()
```
