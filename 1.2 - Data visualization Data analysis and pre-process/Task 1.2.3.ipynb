```python
# При помощи библиотеки pandas импортируйте набор данных об автомобилях в формате CSV. Колонка 'name' должна быть использована в качестве индекса. Десятичный разделитель — точка, разделитель колонок — запятая.

# Подключим библиотеки, импортируем данные и выполним сортировку по колонке hp

import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
all_cars = pd.read_csv('Task 1.2.3 - RAW data auto-mpg-quiz.csv',
                       delimiter=',',
                       index_col='name',
                       decimal='.', 
                       usecols=['mpg','cyl','displ','hp','weight','accel','yr','origin','name'])
```

```python
# Отсортируйте данные по возрастанию мощности автомобилей (колонка hp) и постройте график зависимости ускорения от мощности (accel) от hp))

all_cars = all_cars.sort_values('hp')
```

```python
# Выведем график

plt.plot('hp', 'accel', data=all_cars)
plt.show()
```

```python
# Постройте гистограмму распределения лошадиных сил (столбец hp) среди автомобилей. Используйте метод hist библиотеки matplotlib с параметрами по умолчанию.

# Даные уже импортированы в переменную all_cars. Для построение графика следует указать источник данных и название колонки:
plt.hist('hp', data=all_cars)
plt.show()
```

```python
# Выполните агрегацию данных, определив количество автомобилей с определенным числом цилиндров (столбец cyl). Постройте пирог (метод pie библиотеки matplotlib) на основании полученных данных.

# Данные уже импортированы в переменную all_cars. Для построение пирога выполним группировку по столбцу cyl и подсчитаем количество автомобилей в каждой группе:
all_cars_pie_data = all_cars.groupby(['cyl']).count()
# В каждом из столбцов полученной таблицы содержится количество автомобилей в группе, в индексе указано число цилиндров. Отобразим график:
plt.pie(all_cars_pie_data['mpg'], labels=all_cars_pie_data.index)
plt.show()
```
