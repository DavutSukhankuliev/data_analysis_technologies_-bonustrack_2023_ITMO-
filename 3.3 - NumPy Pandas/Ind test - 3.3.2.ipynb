1. Средствами библиотеки Pandas языка Python считайте данные из файла titanic.csv (набор данных выживших пассажиров Титаника: класс кабины, возраст, пол, порт посадки и т.д.) и найдите требуюмую информацию:

1. Введите количество записей в полученном датафрейме:
2. Введите среднее значение возраста пассажиров (округлите до целого):
3. Сгруппируйте пассажиров по полу и введите среднее значение возраста пассажиров мужского пола (округлите до целого):
4. Введите среднее значение возраста выживших пассажиров женского пола (округлите до целого):
5. Отсортируйте пассажиров по возрасту и введите значение `'PClass'` у пассажира `iloc[1283]` в этом списке:

```python
import pandas as pd

# Считываем данные из файла titanic.csv
df = pd.read_csv('Exercises - 3.3.2 titanic.csv')

# 1. Количество записей в полученном датафрейме
num_records = df.shape[0]
print(f"Количество записей: {num_records}")

# 2. Среднее значение возраста пассажиров (округлённое до целого)
mean_age = round(df['Age'].mean())
print(f"Средний возраст пассажиров: {mean_age}")

# 3. Среднее значение возраста пассажиров мужского пола (округлённое до целого)
mean_age_male = round(df[df['Sex'] == 'male']['Age'].mean())
print(f"Средний возраст пассажиров мужского пола: {mean_age_male}")

# 4. Среднее значение возраста выживших пассажиров женского пола (округлённое до целого)
mean_age_female_survivors = round(df[(df['Sex'] == 'female') & (df['Survived'] == 1)]['Age'].mean())
print(f"Средний возраст выживших пассажиров женского пола: {mean_age_female_survivors}")

# 5. Значение 'PClass' у пассажира iloc[1283] в отсортированном по возрасту списке
sorted_df = df.sort_values(by='Age')
pclass_1283 = sorted_df.iloc[1283]['PClass']
print(f"Значение 'PClass' у пассажира iloc[1283] в отсортированном по возрасту списке: {pclass_1283}")
```

2. Набор данных pulsar_stars  содержит сведения о звездах, полученные в ходе исследовании вселенной (High Time Resolution Universe Survey) с целью определения одного из типа нейтронных звезд — пульсаров. Всего 17898 наблюдений, среди них 1639 — пульсары. 

По каждому наблюдению доступно 8 непрерывных предикторов: среднее значение интегрального профиля; стандартное отклонение интегрального профиля; избыточный эксцесс интегрального профиля; асимметрия интегрального профиля; среднее значение кривой DM-SNR; стандартное отклонение кривой DM-SNR; избыточный эксцесс кривой DM-SNR; асимметрия кривой DM-SNR и бинарный отклик. 

Итак, столбец TARGET (TG) является откликом, все остальные поля — предикторами.

1. Получите выборку из набора данных $\texttt{pulsar_stars}$ на основании следующих критериев:
- Все строки, где $\texttt{TARGET(TG)} =$ `0` и $\texttt{MIP} \in$ `[86.6171875, 87.5859375]`
- И все строки, где $\texttt{TARGET(TG)} =$ `1` и $\texttt{MIP} \in$ `[14.1484375, 20.7890625]`
2. Укажите число записей в полученной выборке:
3. Определите выборочное среднее (среднее в полученной выше выборке) для столбца $\texttt{MIP}$:

```python
import pandas as pd

# Загрузка данных из файла pulsar_stars.csv
df = pd.read_csv('Exercises - 3.3.2 pulsar_stars_new.csv')

# Установка названий столбцов, если необходимо
# col_names = ['MIP', 'STDIP', 'EKIP', 'SIP', 'MC', 'STDC', 'EKC', 'SC', 'TG']
# df.columns = col_names

# 1. Получение выборки на основании критериев
subset_0 = df[(df['TG'] == 0) & (df['MIP'] >= 86.6171875) & (df['MIP'] <= 87.5859375)]
subset_1 = df[(df['TG'] == 1) & (df['MIP'] >= 14.1484375) & (df['MIP'] <= 20.7890625)]

# Объединение выборок
result_subset = pd.concat([subset_0, subset_1])

# 2. Количество записей в полученной выборке
num_records = result_subset.shape[0]
print(f"Количество записей в выборке: {num_records}")

# 3. Выборочное среднее для столбца MIP в полученной выборке
mean_mip = result_subset['MIP'].mean()
print(f"Выборочное среднее для столбца MIP: {mean_mip}")
```
