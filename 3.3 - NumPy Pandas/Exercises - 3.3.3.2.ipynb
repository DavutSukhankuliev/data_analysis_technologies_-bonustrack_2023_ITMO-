```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из CSV файла
df = pd.read_csv('Exercises - 3.3.2 titanic.csv')

# Построение круговой диаграммы распределения возрастов пассажиров
df['Age'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(8, 8))

# Настройка заголовка
plt.title('Distribution of Passengers by Age')

# Показать график
plt.show()
```

2. Средствами библиотеки Pandas языка Python считайте данные из файле titanic.csv (набор данных выживших пассажиров Титаника: класс кабины, возраст, пол, порт посадки и т.д.) и постройте требуемые графики:

Введите строку кода, которая выведет такой график:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из CSV файла
df = pd.read_csv('Exercises - 3.3.2 titanic.csv')

# Построение гистограммы распределения возрастов пассажиров
df['Age'].value_counts().plot.hist(bins=30, edgecolor='black', figsize=(10, 6))

# Настройка подписи осей и заголовка
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Passengers by Age')

# Показать график
plt.show()
```
